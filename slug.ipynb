```python
from slugpy.cluster_slug import cluster_slug
import numpy as np
cs = cluster_slug()
cs.add_filters(['WFC3_UVIS_F336W',
                 'WFC3_UVIS_F438W', 
                 'WFC3_UVIS_F555W',
                'WFC3_UVIS_F814W',
                  'WFC3_UVIS_F657N'])

# Set priors to be flat in log T and A_V, but vary with log M as
# p(log M) ~ 1/M
def priorfunc(physprop):
   # Note: physprop is an array of shape (N, 3) where physprop[:,0] =
   # log M, physprop[:,1] = log T, physprop[:,2] = A_V
   return 1.0/np.exp(physprop[:,0])

cs.priors = priorfunc

logm, pdf = cs.mpdf(0, phot, photerr = photerr)
```
